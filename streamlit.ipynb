preprocessing

casefolding

```python
def casefoldingText(ulasan):
  ulasan = ulasan.lower()
  return ulasan
ulasan['CaseFolding']= ulasan['ulasan'].apply(casefoldingText)
print('Case Folding Result: \n')
print(ulasan['CaseFolding'].head(5))
print('\n\n\n')
```

cleaning

```python
import re
def cleaningulasan(ulasan):
  ulasan = re.sub(r'@[A-Za-a0-9]+',' ',ulasan)
  ulasan = re.sub(r'#[A-Za-z0-9]+',' ',ulasan)
  ulasan = re.sub(r"http\S+",' ',ulasan)
  ulasan = re.sub(r'[0-9]+',' ',ulasan)
  ulasan = re.sub(r"[-()\"#/@;:<>{}'+=~|.!?,_]", " ", ulasan)
  ulasan = ulasan.strip(' ')
  return ulasan
ulasan['Cleaning']= ulasan['CaseFolding'].apply(cleaningulasan)

def clearEmoji(ulasan):
    return ulasan.encode('ascii', 'ignore').decode('ascii')
ulasan['HapusEmoji']= ulasan['Cleaning'].apply(clearEmoji)

def replaceTOM(ulasan):
    pola = re.compile(r'(.)\1{2,}', re.DOTALL)
    return pola.sub(r'\1', ulasan)
ulasan['Spasi']= ulasan['HapusEmoji'].apply(replaceTOM)

print('Cleaning: \n')
print(ulasan['Cleaning'].head(5))
print('\n\n\n')
```

tokenizing

```python
import nltk
from nltk.tokenize import word_tokenize
def tokenizingText(ulasan):
  ulasan = word_tokenize(ulasan)
  return ulasan
ulasan['Tokenizing']= ulasan['HapusEmoji'].apply(tokenizingText)
print('Tokenizing Result: \n')
print(ulasan['Tokenizing'].head(5))
print('\n\n\n')
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import swifter

factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stemmed_wrapper(term):
    return stemmer.stem(term)

term_dict = {}

for document in ulasan['Normalized_Words']:
    for term in document:
        if term not in term_dict:
            term_dict[term] = ' '

for term in term_dict:
    term_dict[term] = stemmed_wrapper(term)
    print(term,":" ,term_dict[term])

def stemmingText(document):
    return [term_dict[term] for term in document]

ulasan['Stemming'] = ulasan['Normalized_Words'].swifter.apply(stemmingText)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.tree import DecisionTreeClassifier
from joblib import dump, load

# Membaca dataset
dataset_path = '/content/drive/MyDrive/SKRIPSI/skripsi (hasil revisi)/labeling/hasil preprocessing baru.csv'
data = pd.read_csv(dataset_path)

# Inisialisasi vectorizer
vectorizer = TfidfVectorizer()

# Mengubah teks menjadi fitur numerik
features = vectorizer.fit_transform(data['ulasan'])

# Menyimpan vectorizer ke file
vectorizer_path = '/content/vectorizer.joblib'
dump(vectorizer, vectorizer_path)

# Inisialisasi model
model = DecisionTreeClassifier()

# Melatih model dengan fitur numerik dan label
model.fit(features, data['label'])

# Menyimpan model ke file
model_path = '/content/decision_tree_model.joblib'
dump(model, model_path)

# Contoh loading vectorizer dan model terpisah
loaded_vectorizer = load(vectorizer_path)
loaded_model = load(model_path)

# Contoh penggunaan vectorizer dan model yang sudah diload
new_text = ["Ini adalah ulasan baru"]
new_features = loaded_vectorizer.transform(new_text)
prediction = loaded_model.predict(new_features)
print(prediction)
```
